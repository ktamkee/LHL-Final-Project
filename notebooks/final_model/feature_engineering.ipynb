### Combine DataFrames

```python
data = pd.concat([EU_mid_game, KR_mid_game, NA_mid_game], axis=0)
data = data.reset_index(drop=True)
data = data.replace({'BLUE': 0, 'RED': 1, 'NONE':2})
print(data.shape)
data.head()     
```

## Feature Engineering

```python
data['TOP_MASTERY_DIFF'] =  data['BTOP_CHAMP_MAST'] - data['RTOP_CHAMP_MAST'] 
data['JGL_MASTERY_DIFF'] =  data['BJGL_CHAMP_MAST'] - data['RJGL_CHAMP_MAST']
data['MID_MASTERY_DIFF'] =  data['BMID_CHAMP_MAST'] - data['RMID_CHAMP_MAST']
data['BOT_MASTERY_DIFF'] =  data['BBOT_CHAMP_MAST'] - data['RBOT_CHAMP_MAST']
data['SUP_MASTERY_DIFF'] =  data['BSUP_CHAMP_MAST'] - data['RSUP_CHAMP_MAST']
data['AVG_LANE_MASTERY_DIFF'] = (data['TOP_MASTERY_DIFF'] + data['JGL_MASTERY_DIFF'] + data['MID_MASTERY_DIFF'] + data['BOT_MASTERY_DIFF'] + data['SUP_MASTERY_DIFF']) / 5
```

```python
data['BSUM_TEAM_MASTERY'] = data['BTOP_CHAMP_MAST'] + data['BJGL_CHAMP_MAST'] + data['BMID_CHAMP_MAST'] + data['BBOT_CHAMP_MAST'] + data['BSUP_CHAMP_MAST']
data['RSUM_TEAM_MASTERY'] = data['RTOP_CHAMP_MAST'] + data['RJGL_CHAMP_MAST'] + data['RMID_CHAMP_MAST'] + data['RBOT_CHAMP_MAST'] + data['RSUP_CHAMP_MAST']
data['TEAM_MASTERY_DIFF'] = data['BSUM_TEAM_MASTERY'] - data['RSUM_TEAM_MASTERY'] 
```

```python
data['B_WLR'] = (data['BTOP_WINS'] + data['BJGL_WINS'] + data['BMID_WINS'] + data['BBOT_WINS'] + data['BSUP_WINS']) / (data['BTOP_LOSSES'] + data['BJGL_LOSSES'] + data['BMID_LOSSES'] + data['BBOT_LOSSES'] + data['BSUP_LOSSES'])
data['R_WLR'] = (data['RTOP_WINS'] + data['RJGL_WINS'] + data['RMID_WINS'] + data['RBOT_WINS'] + data['RSUP_WINS']) / (data['RTOP_LOSSES'] + data['RJGL_LOSSES'] + data['RMID_LOSSES'] + data['RBOT_LOSSES'] + data['RSUP_LOSSES'])
data['TEAM_WLR_DIFF'] = data['B_WLR'] - data['R_WLR']
```

```python
data['TOP_WLR_DIFF'] =  (data['BTOP_WINS'] / data['BTOP_LOSSES']) - (data['RTOP_WINS'] / data['RTOP_LOSSES'])
data['JGL_WLR_DIFF'] =  (data['BJGL_WINS'] / data['BJGL_LOSSES']) - (data['RJGL_WINS'] / data['RJGL_LOSSES'])
data['MID_WLR_DIFF'] =  (data['BMID_WINS'] / data['BMID_LOSSES']) - (data['RMID_WINS'] / data['RMID_LOSSES'])
data['BOT_WLR_DIFF'] =  (data['BBOT_WINS'] / data['BBOT_LOSSES']) - (data['RBOT_WINS'] / data['RBOT_LOSSES'])
data['SUP_WLR_DIFF'] =  (data['BSUP_WINS'] / data['BSUP_LOSSES']) - (data['RSUP_WINS'] / data['RSUP_LOSSES'])
data['AVG_LANE_WLF_DIFF'] = (data['TOP_WLR_DIFF'] + data['JGL_WLR_DIFF'] + data['MID_WLR_DIFF'] + data['BOT_WLR_DIFF'] + data['SUP_WLR_DIFF']) / 5
```

```python
data['BSUM_LP'] = data['BTOP_LEAGUE_POINTS'] + data['BJGL_LEAGUE_POINTS'] + data['BMID_LEAGUE_POINTS'] + data['BBOT_LEAGUE_POINTS'] + data['BSUP_LEAGUE_POINTS']
data['RSUM_LP'] = data['RTOP_LEAGUE_POINTS'] + data['RJGL_LEAGUE_POINTS'] + data['RMID_LEAGUE_POINTS'] + data['RBOT_LEAGUE_POINTS'] + data['RSUP_LEAGUE_POINTS']
data['TEAM_LP_DIFF'] = data['BSUM_LP'] - data['RSUM_LP'] 
```

```python
BFIRST_OBJ = []
RFIRST_OBJ = []

for i in range(len(data)):
    bcounter = 0
    rcounter = 0
    if data.iloc[i]['FIRST_BLOOD'] == 'RED':
        rcounter += 1
    elif data.iloc[i]['FIRST_BLOOD'] == 'BLUE':
        bcounter += 1
    
    if data.iloc[i]['FIRST_DRAG'] == 'RED':
        rcounter += 1
    elif data.iloc[i]['FIRST_DRAG'] == 'BLUE':
        bcounter += 1
    
    if data.iloc[i]['FIRST_HERALD'] == 'RED':
        rcounter += 1
    elif data.iloc[i]['FIRST_HERALD'] == 'BLUE':
        bcounter += 1
    
    if data.iloc[i]['FIRST_TOWER'] == 'RED':
        rcounter += 1
    elif data.iloc[i]['FIRST_TOWER'] == 'BLUE':
        bcounter += 1
    BFIRST_OBJ.append(bcounter)
    RFIRST_OBJ.append(bcounter)
```

```python
data['BFIRST_OBJ_COUNT'] = BFIRST_OBJ
data['RFIRST_OBJ_COUNT'] = RFIRST_OBJ
```

```python
data = data.replace([np.inf, -np.inf], np.nan).dropna(axis=0)
```
